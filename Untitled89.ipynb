```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ucimlrepo import fetch_ucirepo
import numpy as np

# 載入資料
bank = fetch_ucirepo(id=222)
df = pd.concat([bank.data.features, bank.data.targets], axis=1)

# 計算缺失值統計
missing_count = df.isna().sum()
missing_ratio = (missing_count / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing %': missing_ratio
}).sort_values(by='Missing Count', ascending=False)

# 篩選出有缺失的欄位
missing_df_filtered = missing_df[missing_df['Missing Count'] > 0]

# 顯示表格
import plotly.graph_objects as go
from IPython.display import display
display(missing_df_filtered)

# 長條圖視覺化
plt.figure(figsize=(10, 6))
sns.barplot(
    data=missing_df_filtered.reset_index(),
    x='Missing %', y='index', palette='Reds_d'
)
plt.title("Missing Value Percentage by Feature")
plt.xlabel("Missing %")
plt.ylabel("Feature")
plt.grid(True, axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Step 2: 統計 NaN 或 'unknown' 類型的缺失比例
import numpy as np

# 針對常見有 unknown 或 NaN 的欄位
cols_to_check = ['job', 'education', 'contact', 'poutcome']

missing_stats = {}
for col in cols_to_check:
    # 統計 NaN
    nan_ratio = df[col].isna().mean() * 100
    # 若欄位中存在 'unknown' 字串，統計其比例
    if df[col].dtype == 'object':
        unknown_ratio = (df[col] == 'unknown').mean() * 100
    else:
        unknown_ratio = 0
    # 合併為總缺失比例（包含 NaN + 'unknown'）
    total_missing = nan_ratio + unknown_ratio
    missing_stats[col] = {
        'NaN Ratio (%)': round(nan_ratio, 2),
        "'unknown' Ratio (%)": round(unknown_ratio, 2),
        'Total Missing (%)': round(total_missing, 2)
    }

# 顯示結果為表格
missing_df = pd.DataFrame(missing_stats).T
missing_df
```

```python
# Step 3：根據缺失比例刪除高缺失欄位
cols_to_drop = ['contact', 'poutcome']  # 缺失率 > 25%
df = df.drop(columns=cols_to_drop)

print(f"✅ 已移除缺失過高的欄位：{cols_to_drop}")
print(f"📐 新的資料欄位數：{df.shape[1]}")
df.head()
```

```python
# Step 4：將二元類別變數轉為數值
binary_cols = ['default', 'housing', 'loan', 'y']

# 使用 map 函數將 'yes' -> 1, 'no' -> 0
for col in binary_cols:
    df[col] = df[col].map({'yes': 1, 'no': 0})

print("✅ 已完成二元變數轉換（yes/no → 1/0）")
df[binary_cols].head()
```

```python
# Step 5：對多類別變數進行 One-Hot Encoding

# 選取需要做 One-Hot 的欄位
onehot_cols = ['job', 'marital', 'education', 'month', 'day_of_week']

# 使用 pandas 的 get_dummies 做 One-Hot Encoding，不去除第一類別（利於解釋）
df_encoded = pd.get_dummies(df, columns=onehot_cols, drop_first=False)

print("✅ 已完成 One-Hot Encoding，轉換後欄位數：", df_encoded.shape[1])
df_encoded.head()
```

```python
from scipy.stats import ttest_ind
import pandas as pd
import numpy as np

print("📊 Step 7：統計檢定開始")

# 定義要檢定的連續變數欄位
numerical_cols = ['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']

# 建立空的列表來儲存結果
t_results = []

# 開始迴圈進行每個變數的 T 檢定
for col in numerical_cols:
    group1 = df[df['y'] == 1][col].dropna()
    group0 = df[df['y'] == 0][col].dropna()

    n1 = len(group1)
    n0 = len(group0)
    if n1 > 1 and n0 > 1:
        # T 檢定
        t_stat, p_val = ttest_ind(group1, group0, equal_var=False)

        # 計算自由度（Welch's t-test，自由度計算公式）
        s1_sq = np.var(group1, ddof=1)
        s0_sq = np.var(group0, ddof=1)
        df_val = (s1_sq/n1 + s0_sq/n0)**2 / ((s1_sq/n1)**2/(n1-1) + (s0_sq/n0)**2/(n0-1))

        # 儲存結果
        t_results.append({
            'Feature': col,
            't_stat': round(t_stat, 4),
            'df': round(df_val, 2),
            'p_value': round(p_val, 6),
            'mean_0': round(group0.mean(), 2),
            'mean_1': round(group1.mean(), 2),
            'std_0': round(group0.std(), 2),
            'std_1': round(group1.std(), 2),
            'n_0': n0,
            'n_1': n1,
            'Significant': 'Yes' if p_val < 0.05 else 'No'
        })
    else:
        print(f"⚠ {col} | 樣本數不足")

# 將結果轉為 DataFrame 顯示
t_results_df = pd.DataFrame(t_results)
print("\n📋 T-Test Summary Table:")
display(t_results_df)
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

print("📊 Step 7-2：卡方檢定開始")

categorical_cols = ['job', 'marital', 'education', 'month', 'day_of_week']
chi2_results = []

# 進行卡方檢定
for col in categorical_cols:
    if col in df.columns:
        cross_tab = pd.crosstab(df[col], df['y'])
        if cross_tab.shape[0] > 1:
            chi2, p_val, dof, expected = chi2_contingency(cross_tab)

            chi2_results.append({
                'Feature': col,
                'Chi2': round(chi2, 4),
                'df': dof,
                'p_value': round(p_val, 6),
                'Significant': 'Yes' if p_val < 0.05 else 'No',
                'n_categories': cross_tab.shape[0]
            })
        else:
            print(f"⚠ {col} 的類別數不足，無法進行卡方檢定")
    else:
        print(f"⛔ 欄位 {col} 不存在（可能已 One-Hot 處理）")

# 統整為 DataFrame
chi2_df = pd.DataFrame(chi2_results)
print("\n📋 Chi-Square Test Summary Table:")
display(chi2_df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from xgboost import XGBClassifier

# ⬇️ 資料與特徵定義
X = df_encoded.drop('y', axis=1)
y = df_encoded['y']
original_feature_names = X.columns

# ➗ 資料切分
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# ================================
# 🔷 Logistic Regression
# ================================
pipeline_lr = make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
)
pipeline_lr.fit(X_train, y_train)
y_pred_lr = pipeline_lr.predict(X_test)
y_proba_lr = pipeline_lr.predict_proba(X_test)[:, 1]

print("🔹 Logistic Regression")
print("Accuracy:", pipeline_lr.score(X_test, y_test))
print("Classification Report:\n", classification_report(y_test, y_pred_lr, digits=4))

# 混淆矩陣 - Logistic
cm_lr = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues')
plt.title("Logistic Regression - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ROC Curve - Logistic
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_proba_lr)
plt.plot(fpr_lr, tpr_lr, label=f"AUC = {roc_auc_score(y_test, y_proba_lr):.4f}")
plt.plot([0, 1], [0, 1], linestyle='--', color='orange')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Logistic Regression - ROC Curve")
plt.legend()
plt.grid(True)
plt.show()

# 取得 LR 係數
log_coef = pipeline_lr.named_steps['logisticregression'].coef_[0]

# ================================
# 🟣 XGBoost Classifier (替代 Random Forest + SMOTE)
# ================================
# 自動計算 scale_pos_weight = #negative / #positive
neg_count = sum(y_train == 0)
pos_count = sum(y_train == 1)
scale_pos_weight = neg_count / pos_count

xgb_model = XGBClassifier(
    scale_pos_weight=scale_pos_weight,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

# 標準化 + 訓練
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

xgb_model.fit(X_train_scaled, y_train)
y_pred_xgb = xgb_model.predict(X_test_scaled)
y_proba_xgb = xgb_model.predict_proba(X_test_scaled)[:, 1]

print("🔹 XGBoost Classifier")
print("Accuracy:", xgb_model.score(X_test_scaled, y_test))
print("Classification Report:\n", classification_report(y_test, y_pred_xgb, digits=4))

# 混淆矩陣 - XGBoost
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Purples')
plt.title("XGBoost - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ROC Curve - XGBoost
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_proba_xgb)
plt.plot(fpr_xgb, tpr_xgb, label=f"AUC = {roc_auc_score(y_test, y_proba_xgb):.4f}")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("XGBoost - ROC Curve")
plt.legend()
plt.grid(True)
plt.show()

# ================================
# 📊 特徵重要性比較視覺化
# ================================
feature_importance_df = pd.DataFrame({
    'Feature': original_feature_names,
    'Logistic_Coef': log_coef,
    'XGB_Importance': xgb_model.feature_importances_
})

top20 = feature_importance_df.sort_values(by='XGB_Importance', ascending=False).head(20)

plt.figure(figsize=(12, 10))
plt.barh(top20['Feature'][::-1], top20['XGB_Importance'][::-1], color='lightgreen', label='XGBoost')
plt.plot(top20['Logistic_Coef'][::-1], top20['Feature'][::-1], 'ro', label='Logistic Coefficient')
plt.xlabel("Importance / Coefficient")
plt.title("Top 20 Feature Importance Comparison\n(XGBoost vs Logistic Regression)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
